## Importing..

```python
import pandas as pd
import mplfinance as mpf
```

### Reading the Apple stock data, downloaded from:
https://finance.yahoo.com/quote/AAPL/history?p=AAPL

```python
df = pd.read_csv("data/AAPL.csv")
df
```

## Date column to_datetime and setting as index

```python
df.Date = pd.to_datetime(df.Date)
```

```python
df = df.set_index("Date")
```

```python
df
```

## Matplotlib Finance Plots
### .plot() and title, figratio, tight_layout

```python
mpf.plot(df)
```

```python
mpf.plot(df,
        title= "Apple Stock Price 2019/2020",
        figratio=(20,10))
```

```python
mpf.plot(df,
        title= "Apple Stock Price 2019/2020",
        figratio=(20,10),
        tight_layout=True)
```

### Giving a type and showing Volume

```python
mpf.plot(df,
        title= "Apple Stock Price 2019/2020",
        type="line",
        figratio=(20,10),
        volume=True)
```

### Plotting a specific month or an interval

```python
mpf.plot(df['2019-12'],
        title= "Apple Stock Price 2019-Dec",
        type="candle",
        figratio=(20,10),
        tight_layout=True,
        volume=True)
```

```python
mpf.plot(df['2019-12':"2020-03"],
        title= "Apple Stock Price 2019-Dec / 2020-Mar",
        type="line",
        figratio=(20,10),
        tight_layout=True,
        volume=True)
```

### Moving Average

```python
mpf.plot(df['2020-05': '2020-10'],
        title= "Apple Stock Price 2020-May / 2020-Oct",
        type="candle",
        mav=(20),
        figratio=(20,10),
        tight_layout=True,
        volume=True)
```

### Style

```python
mpf.plot(df['2020-05': '2020-10'],
        title= "Apple Stock Price 2020-May / 2020-Oct",
        type="candle",
        mav=(20),
        figratio=(20,10),
        volume=True,
        style="yahoo")
```
